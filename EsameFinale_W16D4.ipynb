```python
import pandas as pd

file_url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
covid = pd.read_csv(file_url, encoding="latin1", sep=",", low_memory=False)
covid
```

### Si richiede di verificare le dimensioni del dataset e le diciture presenti nell'intestazione

```python
print(covid.shape)
print(covid.columns)
```

### Si chiede poi per ogni continente di trovare il numero di casi totali avvenuti in quello stesso continente; si chiede di non considerare eventuali locazioni che nel dataset non appartengono ad alcun continente;

```python
covid_continents = covid.dropna(subset=['continent'])
print("Totale dei casi per ogni continente:", covid_continents.groupby('continent')['total_cases'].sum())
```

### Sempre riguardo i casi di COVID totali, si chiede di sviluppare una funzione che prenda in input il dataset e due nomi di continenti, e che ne confronti i seguenti relativi descrittori statistici: valori minimo e massimo, media, e percentuale rispetto al numero dei casi totali nel mondo (calcolati anche sulle locazioni senza indicazione di continente);

```python
def confronto_descrittori_statistici(covid, cont1, cont2):    
    continenti = [cont1, cont2]
    for continente in continenti:
        filtro = covid[covid.continent == continente]['total_cases']
        percentuale = (filtro.sum()/covid['total_cases'].sum())*100
        print("Descrittori statistici per il continente", continente, ":")
        print("Minimo:", filtro.min(), "\nMassimo:", filtro.max(), "\nMedia:", filtro.mean())
        print("Percentuale rispetto al numero dei casi totali nel mondo:", percentuale, '%\n')
    return 
    

confronto_descrittori_statistici(covid, "Europe", "Oceania")
```

### Si chiede poi di effettuare lo stesso tipo di analisi – anche in questo caso sviluppando una funzione ad hoc – per il numero di vaccinazioni totali per ogni continente

```python
def confronto_vaccinazioni_tot(covid):
    covid_continents = covid.dropna(subset=['continent'])
    continenti = covid_continents['continent'].unique()
    for continente in continenti:
        filtro = covid[covid.continent == continente]['total_vaccinations']
        percentuale = (filtro.sum()/covid['total_vaccinations'].sum())*100
        print("Descrittori statistici per il continente", continente, ":")
        print("Minimo:", filtro.min(), "\nMassimo:", filtro.max(), "\nMedia:", filtro.mean())
        print("Percentuale rispetto al numero di vaccinazioni totali nel mondo:", percentuale, '%\n')
    return 
    

confronto_vaccinazioni_tot(covid)
```
